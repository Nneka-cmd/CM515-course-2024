```python
# 2. load in the GAPDH template strant from the fasta provided here https://github.com/Colorado-State-University-CMB/CM515-course-2024/tree/main/modules/14_Google_Colab
#!git clone https://github.com/Colorado-State-University-CMB/CM515-course-2024.git
path = "/content/CM515-course-2024/modules/14_Google_Colab/"

with open(path + 'gapdh.fasta', 'r') as file:
  file_list = file.readlines()
file_list
```

```python
# 3. calculate the GC content of the strant (report as a percenatage)
def calculate_gc_content(seq):
    gc_count = seq.count('G') + seq.count('C')
    total_bases = len(seq)
    gc_content = (gc_count / total_bases) * 100
    return gc_content

gc_content = calculate_gc_content(file_list[1])
print(gc_content)
```

```python
# 4. Create a function that takes in a dna sequence and finds the reverse compliment of it
def reverse_complement(dna_seq):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    reverse_comp_seq = ''.join([complement[base] for base in reversed(dna_seq)])
    return reverse_comp_seq

print(reverse_complement(file_list[1]))
```

```python
# 5. Convert the GAPDH template strand to the coding strand
def coding_strand(dna_sequence):
    rna_sequence = dna_sequence.replace('T', 'U')
    coding_sequence = rna_sequence.replace('A', 'T').replace('C', 'G').replace('G', 'C')
    return coding_sequence

print(coding_strand(file_list[1]))
```

```python
# 6. convert to the mRNA sequence
def dna_to_mRNA(dna_sequence):
    return dna_sequence.replace('T', 'U')

print(dna_to_mRNA(file_list[1]))

mRNA = dna_to_mRNA(file_list[1])
```

```python
# 7. Translate the mRNA
# make a function that finds the start codon (AUG), then translates the protein until a stop codon is reached.
# stop codons are denoted as "*" in the codon table file provided

def translate_dna(sequence):
  #Create a dictionary representing codons and amino acids https://en.wikipedia.org/wiki/DNA_and_RNA_codon_tables
    genetic_code = {
        'AUA':'I', 'AUC':'I', 'AUU':'I', 'AUG':'M',
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACU':'T',
        'AAC':'N', 'AAU':'N', 'AAA':'K', 'AAG':'K',
        'AGC':'S', 'AGU':'S', 'AGA':'R', 'AGG':'R',
        'CUA':'L', 'CUC':'L', 'CUG':'L', 'CUU':'L',
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCU':'P',
        'CAC':'H', 'CAU':'H', 'CAA':'Q', 'CAG':'Q',
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGU':'R',
        'GUA':'V', 'GUC':'V', 'GUG':'V', 'GUU':'V',
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCU':'A',
        'GAC':'D', 'GAU':'D', 'GAA':'E', 'GAG':'E',
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGU':'G',
        'UCA':'S', 'UCC':'S', 'UCG':'S', 'UCU':'S',
        'UUC':'F', 'UUU':'F', 'UUA':'L', 'UUG':'L',
        'UAC':'Y', 'UAU':'Y', 'UAA':'*', 'UAG':'*',
        'UGC':'C', 'UGU':'C', 'UGA':'*', 'UGG':'W',
    }

    start_codon = 'AUG'
    stop_codons = ['UAA', 'UAG', 'UGA']

    protein = ""
    for i in range(0, len(sequence), 3):
        codon = sequence[i:i+3]
        if codon == start_codon:
            for j in range(i, len(sequence), 3):
                codon = sequence[j:j+3]
                if codon in stop_codons:
                    return protein
                amino_acid = genetic_code.get(codon, '*')
                protein += amino_acid
    return protein
print(translate_dna(mRNA))
```
